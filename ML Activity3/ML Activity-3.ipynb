```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel(r"E:\ML Activities\Activity_3\Activity_3_Data.xlsx")
df = pd.DataFrame(data)
data["Age"] = 2019 - data["Year Built"]
data.set_index("Id")
plt.scatter(data.index,data["Price"])
plt.show()
data
```

# Exercise 3.1

```python
#nomalize
features = ["Quality","Condition","Age","Square Feet","Locality"]

for i in features:
    data[i] = round(( data[i] - data[i].mean() ) / (data[i].max() - data[i].min()),2)
data
```

# Exercise 3.2

```python
theta = np.array([[0],[0],[0],[0],[0],[0]])
X = np.c_[np.ones(100),data["Quality"],data["Condition"],data["Age"],data["Square Feet"],data["Locality"]]
n_iterations = 1000
alpha = 0.1
Y = np.c_[data["Price"]]

def cost_function(theta,X,Y):
    J = np.sum(np.dot((np.dot(X,theta)-Y).T, (np.dot(X,theta)-Y))/ (2*len(Y)))
    return J

cost_history = []
for i in range(n_iterations):
    a = np.sum( theta[0] - alpha * (1/len(Y)) * np.sum(np.dot(X,theta)-Y))
    b = np.sum( theta[1] - alpha * (1/len(Y)) * np.dot(X.T[1],(np.dot(X,theta)-Y)))
    c = np.sum( theta[2] - alpha * (1/len(Y)) * np.dot(X.T[2],(np.dot(X,theta)-Y)))
    d = np.sum( theta[3] - alpha * (1/len(Y)) * np.dot(X.T[3],(np.dot(X,theta)-Y)))
    e = np.sum( theta[4] - alpha * (1/len(Y)) * np.dot(X.T[4],(np.dot(X,theta)-Y)))
    f = np.sum( theta[5] - alpha * (1/len(Y)) * np.dot(X.T[5],(np.dot(X,theta)-Y)))
    theta = np.array([[a],[b],[c],[d],[e],[f]])
    cost_history.append(cost_function(theta,X,Y))

hypo = np.dot(X,theta)
plt.plot(np.linspace(0,1000,1000),cost_history)
plt.title("Cost function")
plt.show()

plt.plot(data.index,hypo,c = "orange",label="Predicted")  
plt.scatter(data.index,data["Price"],c = "g",label = "Original")
plt.legend()
plt.show()
print(theta)
print("The cost is:",cost_function(theta,X,Y))
```

```python
theta1 = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,Y))
print(theta1)
print("The cost for normal equation is:",cost_function(theta1,X,Y))

hypo = np.dot(X,theta1)

plt.plot(data.index,hypo,c = "orange",label="Predicted")  
plt.scatter(data.index,data["Price"],c = "g",label = "Original")
plt.legend()
plt.show()
print(theta)
```

# Regularization

```python
theta3 = np.array([[0],[0],[0],[0],[0],[0]])
X2 = np.c_[np.ones(100),data["Quality"],data["Condition"],data["Age"],data["Square Feet"],data["Locality"]]
n_iterations = 1000
alpha = 0.1
lambd = 10
Y = np.c_[data["Price"]]

def cost_function_regularized(theta,X,Y):
    x = np.sum(np.multiply(theta3,theta3))
    J = sum(np.dot((np.dot(X,theta)-Y).T, (np.dot(X,theta)-Y)), lambd * x )/ (2*len(Y))
    return J

cost3_history = []
for i in range(n_iterations):
    a = np.sum( theta3[0] - alpha * (1/len(Y)) * np.sum(np.dot(X,theta3)-Y))
    b = np.sum( theta3[1] - alpha * (1/len(Y)) * sum(np.dot(X2.T[1],(np.dot(X2,theta3)-Y)) , lambd * theta3[1]))
    c = np.sum( theta3[2] - alpha * (1/len(Y)) * sum(np.dot(X2.T[2],(np.dot(X2,theta3)-Y)) , lambd * theta3[2]))
    d = np.sum( theta3[3] - alpha * (1/len(Y)) * sum(np.dot(X2.T[3],(np.dot(X2,theta3)-Y)) , lambd * theta3[3]))
    e = np.sum( theta3[4] - alpha * (1/len(Y)) * sum(np.dot(X2.T[4],(np.dot(X2,theta3)-Y)) , lambd * theta3[4]))
    f = np.sum( theta3[5] - alpha * (1/len(Y)) * sum(np.dot(X2.T[5],(np.dot(X2,theta3)-Y)) , lambd * theta3[5]))
    theta3 = np.array([[a],[b],[c],[d],[e],[f]])
    cost3_history.append(cost_function_regularized(theta3,X2,Y))
    

hypo2 = np.dot(X2,theta3)
plt.plot(np.linspace(0,1000,1000),cost3_history)
plt.title("Cost function")
plt.show()

plt.plot(data.index,hypo2,c = "orange",label="Predicted")  
plt.scatter(data.index,data["Price"],c = "g",label = "Original")
plt.legend()
plt.show()
print(theta3)
print("The cost after regularisation is:",cost_function(theta3,X2,Y))
```
