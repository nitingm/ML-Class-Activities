```python


data = pd.read_excel(r"E:\ML Activities\activity 1\Activity_1_Data.xlsx")
df = pd.DataFrame(data,columns = ["ID","x_cord","y_cord"])
df
```

```python
plt.figure(figsize = (8,6))
plt.scatter(df['x_cord'],df['y_cord'],label='Customers',c='blue')
plt.scatter(df['x_cord'].mean(),df['y_cord'].mean(),c='y',label = 'Centroid')
plt.scatter(df['x_cord'].mean(),df['y_cord'].mean(),facecolors = 'none' ,edgecolors = 'r',s =29000)
plt.legend()
plt.show()
```

```python
import math
arr = []
brr = []
x0 = df['x_cord'].mean()
y0 = df['y_cord'].mean()
for i in range(len(df)):
    dist = ( (x0-df['x_cord'][i])**2 + (y0 - df['y_cord'][i])**2 )
    if dist <= 4:
        arr.append([df['ID'][i],df['x_cord'][i],df['y_cord'][i]])
    else:
        brr.append([df['ID'][i],df['x_cord'][i],df['y_cord'][i]])

retained_df = pd.DataFrame(arr,columns = ['ID','x_cord','y_cord'])
plt.scatter(retained_df["x_cord"],retained_df["y_cord"])
transferred_df = pd.DataFrame(brr,columns = ['ID','x_cord','y_cord'])
retained_df.to_excel('Activity1-Retain.xlsx')
transferred_df.to_excel('Activity1-Transfer.xlsx')
```
